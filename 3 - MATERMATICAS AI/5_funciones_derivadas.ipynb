# üìå ¬øQU√â ES LA DERIVADA?

# La derivada de una funci√≥n f(x) te dice c√≥mo cambia f cuando x cambia.
# Es como medir la "velocidad" de una funci√≥n.

# ‚úÖ Notaciones comunes:
# f'(x), df/dx, d(f(x))/dx, ‚àÇf/‚àÇx

# üß† ¬øQu√© significa?
# - Si la derivada es positiva ‚Üí la funci√≥n est√° subiendo.
# - Si es negativa ‚Üí la funci√≥n est√° bajando.
# - Si es cero ‚Üí hay un pico, valle o parte plana (punto cr√≠tico).

# üìê En el gr√°fico:
# La derivada es la pendiente de la recta que toca la curva en un solo punto (recta tangente).

# üöó Ejemplo real:
# f(t) = distancia recorrida en el tiempo
# f'(t) = velocidad (cu√°n r√°pido cambia la distancia)

# üîç ¬øPara qu√© sirve?
# - Encontrar m√°ximos o m√≠nimos (ej: optimizar una funci√≥n)
# - Calcular velocidad, aceleraci√≥n en f√≠sica
# - Detectar cambios bruscos en datos o se√±ales

# üõ†Ô∏è En Machine Learning:
# - Las derivadas se usan para entrenar modelos con *descenso del gradiente*

# üìå ¬øQu√© es la derivada?

# La derivada dice c√≥mo cambia una funci√≥n en un punto.
# Es como medir qu√© tan empinada est√° la curva en ese lugar.

# üéØ ¬øC√≥mo se calcula?
# Tom√°s dos puntos cercanos y calcul√°s la pendiente:
# pendiente = (cambio en y) / (cambio en x)

# Luego acerc√°s esos puntos cada vez m√°s ‚Üí eso es un "l√≠mite":
# f'(x) = [f(x + h) - f(x)] / h    cuando h se hace muy chico

# üß† ¬øQu√© significa?
# - Si f'(x) > 0 ‚Üí la funci√≥n sube.
# - Si f'(x) < 0 ‚Üí la funci√≥n baja.
# - Si f'(x) = 0 ‚Üí est√° plana (puede ser un pico o valle).

# üß™ ¬øD√≥nde se usa?
# - En f√≠sica para saber la velocidad.
# - En econom√≠a para ver c√≥mo cambian los precios.
# - En inteligencia artificial para entrenar redes neuronales.

# üìå ¬øTODAS LAS FUNCIONES SON DERIVABLES?

# ‚ùå No todas las funciones tienen derivada en todos sus puntos.
# Para que una funci√≥n tenga derivada en x, debe existir el l√≠mite del cociente incremental en ese punto.
# Porque al acercarse las pendientes deben ser iguales
# Si no coinciden, no hay derivada en ese punto.

# üî¢ Pasos:
# Tomamos dos puntos cercanos: x y ùë•+‚Ñé
# Calculamos la "pendiente" entre ellos: cu√°nto cambia f(x)
# Luego acercamos h a 0, o sea, hacemos que los dos puntos est√©n casi pegados
# Si ese valor de pendiente se estabiliza (llega a un n√∫mero √∫nico), entonces la derivada existe.

# üî∏ Ejemplo cl√°sico:
# f(x) = |x| no es derivable en x = 0.
# Porque al acercarse desde la izquierda y la derecha, las pendientes son distintas ‚Üí no hay un √∫nico valor.
# A la izquierda de 0: la pendiente es -1
# A la derecha de 0: la pendiente es +1

# ‚úÖ Cuando una funci√≥n tiene derivada en todo su dominio, se dice que:
# - Es **derivable** o **suave**
# - Es de **clase C¬π** si adem√°s la derivada es continua

# üß† ¬øPor qu√© importa?
# En c√°lculo, optimizaci√≥n y machine learning, usamos derivadas para:
# - Buscar m√°ximos/m√≠nimos
# - Analizar pendientes
# - Entrenar modelos (gradientes)

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: x**2        # üìå Definimos la funci√≥n f(x) = x¬≤
df = lambda x: 2*x        # üìå Su derivada: f'(x) = 2x

x = np.linspace(-2, 2, 200)  # üî¢ Creamos 200 valores de x entre -2 y 2
x0 = 1                      # üìç Punto donde queremos la tangente
y0 = f(x0)                  # üìà Valor de la funci√≥n en x0 (es decir, f(x0))

# üß† F√≥rmula de la recta tangente: y = f(x0) + f'(x0)(x - x0)
tangente = y0 + df(x0)*(x - x0)

# üñºÔ∏è Graficamos
plt.plot(x, f(x), label='f(x) = x¬≤')             # Curva f(x)
plt.plot(x, tangente, '--', label='Tangente en x=1')  # L√≠nea tangente en x=1
plt.scatter(x0, y0, color='red')                 # Punto rojo donde se toca la tangente
plt.legend(); plt.grid(); plt.show()             # Mostrar leyenda, grilla y gr√°fico
```

## Funci√≥n Sigmoide

La **funci√≥n sigmoide** se define como:

sigma(x) = \frac{1}{1 + e^{-x}}

Es una funci√≥n muy utilizada en *machine learning*, especialmente en:

- **Clasificaci√≥n binaria**: su salida puede interpretarse como una probabilidad (entre 0 y 1).
- **Redes neuronales**: se usa como **funci√≥n de activaci√≥n**, para introducir no linealidades en los modelos.

### üîç Propiedades
- Es **derivable en todo** el conjunto de los n√∫meros reales ‚Ñù.
- Su **derivada tambi√©n es continua**.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool

# Activa la salida de Bokeh en el notebook
output_notebook()

# Definici√≥n de la funci√≥n sigmoidal
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivada de la funci√≥n sigmoidal
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

# Valores de x
x = np.linspace(-10, 10, 400)

# C√°lculo de la funci√≥n sigmoidal y su derivada
y = sigmoid(x)
y_deriv = sigmoid_derivative(x)

# ------------------------ ACA TERMINA LA FUNCION SIGMOIDAL ------------------------

# Crear un DataFrame con los datos
data = pd.DataFrame({
    'x': x,
    'Sigmoid': y,
    'Sigmoid Derivative': y_deriv
})


# Crear la figura de Bokeh
p = figure(title="Sigmoid Function and its Derivative",
           x_axis_label='x', y_axis_label='y',
           width=800, height=400)

# Agregar la funci√≥n sigmoidal
p.line(data['x'], data['Sigmoid'], legend_label="Sigmoid Function", line_width=2)

# Agregar la derivada de la funci√≥n sigmoidal
p.line(data['x'], data['Sigmoid Derivative'], legend_label="Sigmoid Derivative", line_width=2, line_dash='dashed', color='orange')

# Agregar herramientas de hover
hover = HoverTool()
hover.tooltips = [("x", "@x"), ("Sigmoid", "@Sigmoid"), ("Sigmoid Derivative", "@Sigmoid Derivative")]
p.add_tools(hover)

# Mostrar la gr√°fica
show(p)
```

# üìå 1. DERIVACI√ìN ANAL√çTICA

# Cuando derivamos anal√≠ticamente, buscamos una f√≥rmula exacta para la derivada de una funci√≥n.
# En lugar de usar el cociente incremental cada vez, usamos reglas conocidas para derivar de forma directa.

# ‚úÖ TABLA B√ÅSICA DE DERIVADAS

# Constante:        f(x) = c                   ‚Üí  f'(x) = 0
# Suma:             f(x) = g(x) + h(x)         ‚Üí  f'(x) = g'(x) + h'(x)
# Producto:         f(x) = g(x)¬∑h(x)           ‚Üí  f'(x) = g¬∑h' + g'¬∑h
# Cociente:         f(x) = g(x)/h(x)           ‚Üí  f'(x) = (g'¬∑h ‚àí g¬∑h') / h¬≤
# Potencia:         f(x) = x^r (r ‚â† 0)         ‚Üí  f'(x) = r¬∑x^(r‚àí1)
# Exponencial:      f(x) = exp(x)              ‚Üí  f'(x) = exp(x)
# Logaritmo:        f(x) = ln(x)               ‚Üí  f'(x) = 1/x
# Seno:             f(x) = sin(x)              ‚Üí  f'(x) = cos(x)
# Coseno:           f(x) = cos(x)              ‚Üí  f'(x) = ‚àísin(x)
# Tangente:         f(x) = tan(x)              ‚Üí  f'(x) = 1/cos¬≤(x)
# Regla de la cadena: f(x) = g(h(x))           ‚Üí  f'(x) = g'(h(x))¬∑h'(x)

# üì¶ Ejemplo:
# Si f(x) = x¬≥, usamos la regla de potencia:
# f'(x) = 3¬∑x¬≤

# üí° Tip:

# üìå 2.1.1 - REGLA DEL PRODUCTO

# Si f(x) es un producto de funciones: f(x) = u(x) ¬∑ v(x)
# Entonces su derivada se calcula con la regla del producto:

# df/dx = v(x) ¬∑ du/dx + u(x) ¬∑ dv/dx

# üî¨ Ejemplo aplicado a neurociencia:
# La funci√≥n alfa del potencial postsin√°ptico se define como:
# f(t) = t ¬∑ exp(-t / œÑ)
# donde œÑ (tau) es una constante temporal.

# Usamos la regla del producto:
# u(t) = t
# v(t) = exp(-t / œÑ)

```python
import sympy as sp

# Definimos la variable
x = sp.symbols('x')

# Definimos la funci√≥n como producto: f(x) = x * (x + 1)
u = x
v = x + 1
f = u * v

# Calculamos la derivada usando la regla del producto
f_deriv = sp.diff(f, x)

# Mostramos resultados
print("Funci√≥n original: ", f)
print("Derivada: ", f_deriv)
```

```python
# üìå 2.2.3 - DERIVADAS DE ORDEN SUPERIOR
# Pod√©s derivar una funci√≥n m√°s de una vez.
# Segunda derivada: d¬≤f/dx¬≤ ‚Üí mide la concavidad (si se acelera o desacelera)
# En SymPy se hace con: sp.diff(f, x, 2)  o  sp.Derivative(f, x, x)

# üìå 2.2.4 - DERIVACI√ìN SIMB√ìLICA EN PYTHON: SYMPY

# SymPy permite crear funciones simb√≥licas y derivarlas f√°cilmente.
# Tambi√©n pod√©s evaluar, graficar, integrar o simplificar expresiones simb√≥licas.


import numpy as np
import matplotlib.pyplot as plt

# 1. Definir s√≠mbolo simb√≥lico (variable)
t = sp.Symbol('t')

# 2. Definir una funci√≥n sigmoide como ejemplo
f = 1 / (1 + sp.exp(-(t - 5)))  # Funci√≥n suave usada en ML

# 3. Derivar simb√≥licamente (1ra derivada)
df = sp.diff(f)
print('Primera derivada simb√≥lica:', df)

# 4. Otra forma: usar Derivative (se eval√∫a luego con .doit())
deriv = sp.Derivative(f, t)
print('Sin evaluar:', deriv)
print('Evaluado con .doit():', deriv.doit())

# 5. Segunda derivada (orden 2)
deriv_2 = sp.Derivative(f, t, t).doit()
print('Segunda derivada:', deriv_2)

# 6. Evaluar la derivada en valores usando numpy
df_np = sp.lambdify(t, df, 'numpy')  # convierte a funci√≥n normal

print('Evaluaci√≥n en t=5:', df_np(5))  # valor puntual
print('Evaluaci√≥n en array:', df_np(np.arange(0, 10)))  # array de valores
```

```python
# üî¨ EJERCICIO DE PROGRAMACI√ìN 2.2.4
# Derivar simb√≥licamente el potencial postsin√°ptico del ejercicio 2.1.2
# f(t) = E_syn ¬∑ (t - t‚ÇÄ) ¬∑ exp(-(t - t‚ÇÄ)/œÑ)

# Declarar los s√≠mbolos
t, tau, E_syn, t_0 = sp.symbols('t tau E_syn t_0')

# Definir la funci√≥n PSP (potencial postsin√°ptico)
psp = E_syn * (t - t_0) * sp.exp(-(t - t_0) / tau)

# Derivar con respecto a t
diff_psp = sp.diff(psp, t)
print('Derivada del PSP:', diff_psp)

# ‚úîÔ∏è Listo para evaluar o graficar si se asignan valores num√©ricos a las variables
```

```python
# üìå 3.1.2 - EL GRADIENTE

# El gradiente ‚àáf(x‚Éó) es un vector con todas las derivadas parciales de una funci√≥n multivariable f(x‚ÇÅ, x‚ÇÇ, ..., x‚Çô).
# Es decir, mide c√≥mo cambia f en cada direcci√≥n del espacio.

# üìê Se calcula as√≠:
# ‚àáf(x‚Éó‚Çê) = [ ‚àÇf/‚àÇx‚ÇÅ , ‚àÇf/‚àÇx‚ÇÇ , ..., ‚àÇf/‚àÇx‚Çô ]·µÄ

# En Python usamos SymPy para calcularlo f√°cilmente.

import sympy as sp

# Definir los s√≠mbolos
x1, x2, x3 = sp.symbols('x1 x2 x3')

# Definir la funci√≥n multivariable
f = 2*x1*x2 - sp.exp(x3) + x1*x3**4

# Derivadas parciales (gradiente)
df_dx1 = sp.diff(f, x1)
df_dx2 = sp.diff(f, x2)
df_dx3 = sp.diff(f, x3)

# Mostrar cada componente
print("‚àÇf/‚àÇx1 =", df_dx1)
print("‚àÇf/‚àÇx2 =", df_dx2)
print("‚àÇf/‚àÇx3 =", df_dx3)

# Vector gradiente como tupla o matriz columna
gradient = sp.Matrix([df_dx1, df_dx2, df_dx3])
print("\nGradiente ‚àáf(x):")
sp.pprint(gradient)
```

```python
# üìå 4 - C√ÅLCULO INTEGRAL

# El c√°lculo integral responde a: ¬øCu√°l es el √°rea bajo una curva?
# Se relaciona con la derivada a trav√©s del Teorema Fundamental del C√°lculo:
# Si F(x) = ‚à´‚ÇÄÀ£ f(t) dt, entonces F‚Ä≤(x) = f(x)

# üìê La integral definida se expresa como:
# ‚à´‚Çê·µá f(x) dx ‚Üí √°rea bajo la curva entre x = a y x = b

# Para calcularlo num√©ricamente usamos sumas de Riemann o m√©todos como `quad` de SciPy

# üîß EJEMPLO: √Årea bajo sin(x) entre 0 y œÄ

import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from scipy.integrate import quad

# Definir la funci√≥n a integrar
def f(x):
    return np.sin(x)

# L√≠mites de integraci√≥n
a, b = 0, np.pi

# Calcular √°rea bajo la curva con integraci√≥n num√©rica
area, _ = quad(f, a, b)

# Crear datos para graficar
x_vals = np.linspace(0, 2 * np.pi, 500)
y_vals = f(x_vals)

# Gr√°fico de la funci√≥n
fig = px.line(x=x_vals, y=y_vals, labels={'x': 'x', 'y': 'sin(x)'},
              title='√Årea bajo la curva sin(x) de 0 a œÄ')

# Rellenar el √°rea bajo la curva entre a y b
fig.add_trace(go.Scatter(
    x=np.linspace(a, b, 500),
    y=f(np.linspace(a, b, 500)),
    fill='tozeroy',
    mode='none',
    name='√Årea bajo sin(x)'
))

# Anotar valor del √°rea
fig.add_annotation(
    x=np.pi / 2,
    y=0.5,
    text=f"√Årea = {area:.2f}",
    showarrow=True,
    arrowhead=2
)

# Mostrar gr√°fico interactivo
fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go
from scipy.integrate import dblquad

# Definici√≥n de la funci√≥n
def f(x, y):
    return x**2 + y**2

# L√≠mites de integraci√≥n
a, b = 0, 1
c, d = 0, 1

# C√°lculo de la integral doble
volume, _ = dblquad(f, a, b, lambda x: c, lambda x: d) #recordar que solo nos interesa el valor de la integral y no su error por ello el _

# Generaci√≥n de datos para la superficie
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
x, y = np.meshgrid(x, y)
z = f(x, y)

# Crear la figura con Plotly
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# A√±adir un t√≠tulo y ajustar las etiquetas
fig.update_layout(
    title=f'Volumen bajo la superficie x^2 + y^2, Volumen = {volume:.2f}',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
    )
)

# Mostrar la gr√°fica
fig.show()
```
