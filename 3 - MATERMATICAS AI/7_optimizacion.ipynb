# üìå Optimizaci√≥n en Matem√°ticas

## üéØ ¬øQu√© es?
La optimizaci√≥n es el proceso de encontrar los **m√°ximos o m√≠nimos** de una funci√≥n, llamada **funci√≥n objetivo**, posiblemente sujeta a **restricciones**.

## üî¢ Funci√≥n Objetivo
- Es la funci√≥n que se quiere maximizar o minimizar.
- Notaci√≥n com√∫n:  
  f(mathbf{x}) = f(x_1, x_2, \ldots, x_n)

  donde ( \mathbf{x} ) son las **variables de decisi√≥n**.

## üß† ¬øC√≥mo se encuentran m√°ximos y m√≠nimos?
- Se utilizan **derivadas** para hallar:
  - **Puntos cr√≠ticos**: donde la derivada se anula.
  - **Puntos en los bordes del dominio**.
- Luego se analiza si se trata de un **m√°ximo**, **m√≠nimo** o **punto de inflexi√≥n**.

## üõ†Ô∏è Proceso de Optimizaci√≥n
- M√©todos comunes:
  - **Anal√≠ticos**: con c√°lculo diferencial.
  - **Num√©ricos**: como el **gradiente descendente** (utilizado en machine learning).

# üìå Optimizaci√≥n en Deep Learning

## üéØ ¬øPor qu√© es importante?
La **optimizaci√≥n** es clave en deep learning y machine learning, ya que permite que los modelos **aprendan** ajustando sus par√°metros para realizar buenas predicciones.

## ‚öôÔ∏è Ajuste de Par√°metros
Los modelos de deep learning tienen **muchos par√°metros** (pesos) que deben ajustarse para minimizar el error en las predicciones.  
üëâ La optimizaci√≥n busca los **valores √≥ptimos** de esos par√°metros.

## üß† Redes Neuronales
- Las redes ajustan los **pesos sin√°pticos** entre neuronas.
- Usan **descenso de gradiente** u otros m√©todos para minimizar la **funci√≥n de p√©rdida** y mejorar la precisi√≥n.

## üìö Aprendizaje Supervisado
- El modelo aprende a mapear entradas a salidas conocidas.
- La optimizaci√≥n ajusta el modelo para que las predicciones **se acerquen lo m√°s posible** a los valores reales.

## üß± Aprendizaje Profundo
- Los modelos profundos tienen **muchas capas**.
- La optimizaci√≥n es vital para que **todas las capas** contribuyan correctamente al aprendizaje.

## üß© Consideraciones Clave
- Seleccionar algoritmos de optimizaci√≥n eficientes.
- Manejar **problemas comunes** como:
  - **Overfitting** (sobreajuste)
  - **M√≠nimos locales**
  - **Desvanecimiento/explosi√≥n del gradiente**

‚úÖ En resumen:  
La optimizaci√≥n permite que los modelos **aprendan de los datos**, ajustando sus par√°metros para **generalizar** correctamente a nuevos ejemplos.

# üöó Ejemplo Num√©rico de Optimizaci√≥n en la Vida Cotidiana

## üéØ Objetivo
Minimizar el **tiempo total de viaje** entre tu casa y el trabajo.

## üìê Datos del problema
- üìè Distancia fija: `d = 20 km`
- üèéÔ∏è Velocidad variable: `x` (km/h)
- üîí Restricciones:
  - `x ‚â• 0` (no se puede ir en reversa)
  - `x ‚â§ 100` (velocidad m√°xima permitida)

## üß† Funci√≥n Objetivo

Queremos minimizar el tiempo total de viaje:

f(x) = \frac{20}{x}

Donde:
- \( d = 20 \) es la distancia en km
- \( x \) es la velocidad en km/h

## üìâ Derivada de la Funci√≥n

Calculamos la derivada para encontrar extremos:

f'(x) = -\frac{20}{x^2}

Buscamos puntos cr√≠ticos igualando la derivada a cero:

-\frac{20}{x^2} = 0

‚õî Esto **no tiene soluci√≥n real**, ya que la derivada nunca se anula (la funci√≥n siempre decrece en su dominio v√°lido).

## üß™ Evaluaci√≥n en los extremos

Como no hay extremos interiores, evaluamos los **bordes del intervalo permitido** \([0, 100]\):

- \( f(0) \) no est√° definido (divisi√≥n por cero)
- \( f(100) = \frac{20}{100} = 0.2 \) horas = **12 minutos**

## ‚úÖ Conclusi√≥n

La **velocidad √≥ptima** que **minimiza el tiempo** dentro del intervalo permitido es:

x = 100 \, \text{km/h} \Rightarrow t = 12 \, \text{minutos}
Esto demuestra c√≥mo el an√°lisis de derivadas y extremos nos ayuda a resolver problemas cotidianos con herramientas de **optimizaci√≥n matem√°tica**.

```python
# Ejercicio 1: Minimizaci√≥n de P√©rdidas
# Una empresa produce (x) unidades de un art√≠culo y desea minimizar sus p√©rdidas. La funci√≥n de p√©rdida est√° dada por:

# [ L(x) = 3x^2 + 120x + 500 ]

# Encuentra el valor de (x) que minimiza las p√©rdidas y comenta sobre la interpretaci√≥n del resultado en el contexto del problema.

import sympy as sp

# üìå Variable simb√≥lica
x = sp.symbols('x')

# üìâ Funci√≥n de p√©rdida
L = 3*x**2 + 120*x + 500

# üßÆ Derivada (para encontrar extremos)
L_deriv = sp.diff(L, x)

# üîç Puntos cr√≠ticos: donde L' = 0
crit_points = sp.solve(L_deriv, x)

# üß™ Segunda derivada (clasificaci√≥n)
L_second_deriv = sp.diff(L_deriv, x)

# ‚úÖ Resultado
optimal_x = crit_points[0]
clasificacion = "m√≠nimo local" if L_second_deriv > 0 else "m√°ximo o indefinido"

# üñ®Ô∏è Resultados
print("Funci√≥n de p√©rdida:", L)
print("Derivada:", L_deriv)
print("Punto cr√≠tico:", optimal_x)
print("Segunda derivada:", L_second_deriv)
print("Clasificaci√≥n:", clasificacion)

# La segunda derivada se utiliza para determinar la concavidad de la funci√≥n y, por lo tanto, identificar si el punto cr√≠tico encontrado es un m√≠nimo local, 
# un m√°ximo local o un punto de inflexi√≥n. En este caso, la segunda derivada es positiva (6), lo que indica que la funci√≥n de p√©rdida es c√≥ncava hacia arriba 
# en (x = -20). Esto confirma que (x = -20) corresponde a un m√≠nimo local de la funci√≥n de p√©rdida.

# El hecho de que ùë• = -20 sea el valor que minimiza las p√©rdidas significa que, seg√∫n el modelo matem√°tico, las p√©rdidas ser√°n m√°s bajas 
# cuando la empresa no produzca ninguna unidad (es decir, ùë• = 0). Esto puede interpretarse como una se√±al de que la empresa podr√≠a estar 
# operando en un punto donde la producci√≥n actual est√° generando m√°s p√©rdidas de lo necesario, y que ser√≠a beneficioso administrar mejor producci√≥n. üìâüè≠üíº
```

```python
# Ejercicio 2: Maximizaci√≥n de Ganancias
# Una tienda vende (y) productos y busca maximizar sus ganancias. La funci√≥n de ganancias est√° modelada por:

# G(y) = 15y - 0.2y^2

# Utiliza derivadas para encontrar el valor √≥ptimo de (y) y comenta sobre la interpretaci√≥n econ√≥mica del resultado.

import sympy as sp

# Definici√≥n de la variable y la funci√≥n de ganancias
y = sp.symbols('y')
profit_function = 15*y - 0.2*y**2

# Derivada de la funci√≥n de ganancias
derivative_profit = sp.diff(profit_function, y)
print("Derivada de la funci√≥n de ganancias:", derivative_profit)

# Puntos cr√≠ticos (donde la derivada es cero)
critical_points_profit = sp.solve(derivative_profit, y)
print("Puntos cr√≠ticos de la funci√≥n de ganancias:", critical_points_profit)

# Segunda derivada para determinar si es un m√°ximo local
second_derivative_profit = sp.diff(derivative_profit, y)
print("Segunda derivada de la funci√≥n de ganancias:", second_derivative_profit)

# Seleccionamos el punto cr√≠tico donde la concavidad es negativa (m√°ximo local)
optimal_y = critical_points_profit[0]  # Como la concavidad es negativa, es el m√°ximo local
print(f"El valor de y que maximiza las ganancias es: {optimal_y}")

# La segunda derivada se utiliza para determinar la concavidad de la funci√≥n y, por lo tanto, identificar si el punto cr√≠tico encontrado es un m√≠nimo local, 
# un m√°ximo local o un punto de inflexi√≥n. En este caso, la segunda derivada es negativa (-0.4), lo que indica que la 
# funci√≥n de ganancias es c√≥ncava hacia abajo en (x = 37.5). Esto confirma que (x = 37.5) corresponde a un m√°ximo local en la funci√≥n de ganancias.

# Sustituir el valor en la funci√≥n original de ganancias
profit_at_optimal_y = profit_function.subs(y, optimal_y)

print(f"Ganancia en y = {optimal_y}: {profit_at_optimal_y}")

# La ganancia en (y = 37.5) es (281.25). Este resultado se obtuvo al maximizar la funci√≥n de ganancias (G(y) = 15y - 0.2y^2) utilizando derivadas. 
# La interpretaci√≥n econ√≥mica es que la tienda maximiza sus ganancias cuando vende aproximadamente (37.5) productos, y en ese caso, la ganancia alcanza (281.25).

# La segunda derivada en (y = 37.5) es (-0.4), lo cual indica que la funci√≥n de ganancias (G(y) = 15y - 0.2y^2) es c√≥ncava hacia abajo en ese punto. 
# Dado que la segunda derivada es negativa, esto confirma que (y = 37.5) es un m√°ximo local de la funci√≥n de ganancias.
```

```python

import sympy as sp  # material para revisar en autonomia

# Definimos la variable y la nueva funci√≥n
y = sp.symbols('y')
profit_function = y**3 - 6*y**2 + 9*y + 1

# Derivada de la funci√≥n
derivative_profit = sp.diff(profit_function, y)
print("Derivada de la funci√≥n:", derivative_profit)

# Puntos cr√≠ticos (donde la derivada es cero)
critical_points = sp.solve(derivative_profit, y)
print("Puntos cr√≠ticos:", critical_points)

# Segunda derivada para clasificar los puntos
second_derivative = sp.diff(derivative_profit, y)
print("Segunda derivada:", second_derivative)

# Clasificamos los puntos cr√≠ticos
for point in critical_points:
    concavity = second_derivative.subs(y, point)
    if concavity < 0:
        print(f"En y = {point} hay un M√ÅXIMO local.")
    elif concavity > 0:
        print(f"En y = {point} hay un M√çNIMO local.")
    else:
        print(f"En y = {point} la prueba de la segunda derivada es inconclusa.")
```

```python
# Ejercicio 4: (como tarea) Se tiene un conjunto de datos que incluye el tiempo que un usuario pasa en un sitio web y si realiza o no una compra (1 si realiza una compra, 0 si no). 
# La probabilidad de realizar una compra se modela con la funci√≥n sigmoide (P(t) = \frac{1}{1 + e^{-mt}}), 
# donde (t) es el tiempo en minutos y (m) es el par√°metro a ajustar.

# Objetivo: Encontrar el valor de (m) que maximiza la probabilidad de compra, es decir, que minimiza la funci√≥n de p√©rdida asociada a la regresi√≥n log√≠stica.

# Preguntas:

# Define la funci√≥n de p√©rdida para la regresi√≥n log√≠stica.
# Utiliza m√©todos de optimizaci√≥n para encontrar el valor √≥ptimo de (m).

import numpy as np
from scipy.optimize import minimize

# Datos de ejemplo (tiempo en minutos y compras)
time_data = np.array([10, 20, 30, 40, 50])
purchase_data = np.array([0, 0, 1, 1, 1])

# Funci√≥n de p√©rdida para regresi√≥n log√≠stica
def logistic_loss(m):
    predictions = 1 / (1 + np.exp(-m * time_data))  # Calcula las predicciones de compra
    predictions = np.clip(predictions, 1e-15, 1 - 1e-15)  # Ajusta las predicciones para evitar log(0) y log(1)
    return -np.sum(purchase_data * np.log(predictions) + (1 - purchase_data) * np.log(1 - predictions))  # Calcula la p√©rdida log√≠stica

# Encontrar el valor de m que minimiza la p√©rdida
optimal_m_logistic = minimize(logistic_loss, x0=[1], method='L-BFGS-B').x[0]  # Encuentra el valor √≥ptimo de m mediante minimizaci√≥n
#`x0=[1]` establece el punto de partida para la b√∫squeda del m√≠nimo de la funci√≥n de p√©rdida,
#y `.x[0]` extrae el valor √≥ptimo de \(m\) encontrado por el algoritmo de optimizaci√≥n.


# Mostrar resultado
print("\nEjercicio 6: Ajuste de Probabilidad con Regresi√≥n Log√≠stica")
print(f"Valor de m que maximiza la probabilidad de compra: {optimal_m_logistic}")  # Muestra el valor √≥ptimo de m
```

```python
# La funci√≥n minimize y la funci√≥n minimize_scalar son parte del paquete scipy.optimize, pero tienen diferentes prop√≥sitos y aplicaciones.

# minimize: Esta funci√≥n se utiliza para minimizar una funci√≥n de varias variables (es decir, una funci√≥n que toma varios valores como entrada 
# y devuelve un solo valor como salida). Puede manejar problemas de optimizaci√≥n multidimensionales y ofrece una variedad de algoritmos de optimizaci√≥n 
# que se adaptan a diferentes tipos de funciones y restricciones.

# minimize_scalar: Esta funci√≥n, como su nombre indica, se utiliza para minimizar una funci√≥n de una sola variable (es decir, una funci√≥n que toma un 
# solo valor como entrada y devuelve un solo valor como salida). Est√° dise√±ada espec√≠ficamente para problemas de optimizaci√≥n unidimensionales y utiliza 
# m√©todos que son eficientes para este tipo de problemas. por ejemplo calculos de funciones cuadraticas, exponenciales y/o lineales

from scipy.optimize import minimize_scalar

# Definir la funci√≥n cuadr√°tica como una funci√≥n de Python
def funcion_cuadratica(x):
    return x**2 - 6*x + 5

# Minimizar la funci√≥n cuadr√°tica utilizando minimize_scalar
resultado_minimizacion = minimize_scalar(funcion_cuadratica)

# Imprimir el resultado
print("M√≠nimo de la funci√≥n cuadr√°tica:", resultado_minimizacion.x)
```

```python
from scipy.optimize import minimize_scalar

# Definimos la funci√≥n
def profit(y):
    return -(y**3 - 6*y**2 + 9*y + 1)  # Negativo para buscar el m√°ximo

# Buscamos el m√≠nimo de la funci√≥n negativa
result = minimize_scalar(profit)

# El m√°ximo real est√° en el mismo punto pero con signo opuesto
print("y que maximiza la funci√≥n:", result.x)
print("M√°ximo local (valor de la funci√≥n):", -result.fun)

# ¬øPor qu√© minimize_scalar encontr√≥ solo un punto?
# Porque minimize_scalar es un m√©todo num√©rico que solo encuentra un extremo local (m√≠nimo o m√°ximo), dependiendo del intervalo de b√∫squeda.
# Por defecto, intenta buscar globalmente, pero puede quedar atrapado en un solo punto.

# üß† En este caso:
# Como buscaste el m√≠nimo de la funci√≥n negativa, lo que encontraste fue el m√°ximo local en y = 1.
# Pero a diferencia de sympy, no te devuelve todos los puntos cr√≠ticos, solo uno.
# üéØ Resumen: diferencias clave
# üõ†Ô∏è M√©todo	üîç Encuentra todos los extremos	üßÆ Usa derivadas simb√≥licas	üéØ Resultado depende del punto inicial
# sympy	‚úÖ S√≠	‚úÖ S√≠	‚ùå No
# scipy.optimize.minimize_scalar	‚ùå No (solo uno)	‚ùå No (usa m√©todos num√©ricos)	‚úÖ S√≠
# ‚úÖ ¬øCu√°l usar?
# Si quer√©s encontrar todos los extremos (m√≠nimos y m√°ximos locales): sympy es ideal.
# Si solo necesit√°s uno (por ejemplo, el m√°ximo m√°s alto) y est√°s trabajando con datos reales o funciones m√°s complejas: scipy es m√°s pr√°ctico.
```
