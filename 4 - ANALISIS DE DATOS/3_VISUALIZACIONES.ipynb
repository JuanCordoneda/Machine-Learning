

# Importancia de Crear Visualizaciones Atractivas con Python

Las visualizaciones son una herramienta esencial en el análisis de datos y la comunicación de resultados. En Python, existen diversas bibliotecas especializadas, como Matplotlib, Plotly, y otras, que permiten crear visualizaciones atractivas e informativas.

## **¿Por qué son importantes las visualizaciones?**

Las visualizaciones desempeñan un papel crucial en la comprensión de datos y la toma de decisiones informadas. Aquí hay algunas razones clave para prestar atención a la calidad de las visualizaciones:

1. **Comprensión rápida:** Las visualizaciones proporcionan una forma rápida y efectiva de entender patrones, tendencias y relaciones en los datos.

2. **Comunicación efectiva:** Las imágenes son poderosas. Una visualización bien diseñada puede comunicar de manera más efectiva la historia que subyace en los datos, facilitando la transmisión de información a colegas y audiencias no técnicas.

3. **Identificación de patrones:** Las visualizaciones permiten identificar patrones y anomalías que podrían pasar desapercibidos en los datos crudos.

4. **Soporte para la toma de decisiones:** Las decisiones informadas se benefician de una comprensión clara de los datos. Las visualizaciones ofrecen una visión rápida y efectiva para respaldar el proceso de toma de decisiones.

## **Bibliotecas populares de visualización en Python**

1. **Matplotlib:** Una biblioteca versátil y ampliamente utilizada para crear gráficos 2D y 3D en Python.

2. **Plotly:** Destaca por sus capacidades de visualización interactiva, siendo ideal para presentaciones y aplicaciones web.

3. **Seaborn:** Construida sobre Matplotlib, Seaborn simplifica la creación de visualizaciones atractivas con menos líneas de código.

***Matplotlib:***

mpl_toolkits.mplot3d: Matplotlib es una biblioteca de trazado 2D que también incluye capacidades 3D a través del módulo mpl_toolkits.mplot3d. Es fácil de usar y es una opción sólida para gráficos 3D básicos.

```python
# Importar la biblioteca de trazado de Matplotlib y el módulo 3D
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Importar la biblioteca NumPy para generación de datos
import numpy as np

# Generar datos de ejemplo
np.random.seed(42)  # Establecer una semilla para reproducibilidad
n_points = 100
x = np.random.rand(n_points)
y = np.random.rand(n_points)
z = np.random.rand(n_points)

# Crear la figura 3D
fig = plt.figure(figsize=(8, 6))

# Agregar un subplot 3D a la figura
ax = fig.add_subplot(111, projection='3d')

# Crear un gráfico de dispersión 3D
ax.scatter(x, y, z, c='blue', marker='p') #s / #p

# Configuración adicional del gráfico
ax.set_xlabel('X Label')  # Etiqueta del eje X
ax.set_ylabel('Y Label')  # Etiqueta del eje Y
ax.set_zlabel('Z Label')  # Etiqueta del eje Z
ax.set_title('Scatter Plot 3D')  # Título del gráfico

# Mostrar el gráfico
plt.show()

```

**Plotly:**

plotly.graph_objects: Plotly es conocido por su capacidad para crear gráficos interactivos. La biblioteca plotly.graph_objects permite crear gráficos 3D interactivos y es especialmente útil para la visualización en línea.

```python
import plotly.graph_objects as go
import numpy as np

# Generar datos de ejemplo
np.random.seed(42)
n_points = 100
x = np.random.rand(n_points)
y = np.random.rand(n_points)
z = np.random.rand(n_points)

# Crear el gráfico de dispersión 3D interactivo
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='markers', marker=dict(color='blue', size=8))]) #lines+markers / text

# Configuración adicional del gráfico
fig.update_layout(scene=dict(xaxis_title='X Label', yaxis_title='Y Label', zaxis_title='Z Label'),
                  title='Scatter Plot 3D - Plotly')

# Mostrar el gráfico
fig.show()
```

```python
import plotly.graph_objects as go  # Importa la clase 'graph_objects' de Plotly y la renombra como 'go'
import numpy as np

# Crear una malla de datos 3D
x = np.linspace(-5, 5, 100)  # Genera 100 puntos espaciados uniformemente en el rango de -5 a 5 para el eje x
y = np.linspace(-5, 5, 100)  # Genera 100 puntos espaciados uniformemente en el rango de -5 a 5 para el eje y
x, y = np.meshgrid(x, y)  # Crea una cuadrícula 2D a partir de los puntos x e y
z = np.sin(np.sqrt(x**2 + y**2))  # Calcula el valor de z para cada punto de la cuadrícula usando la función seno

# Crear el gráfico de superficie 3D interactivo
fig = go.Figure()  # Crea una instancia de la clase 'Figure' de Plotly

fig.add_trace(go.Surface(z=z, x=x, y=y,  # Agrega un gráfico de superficie 3D a la figura
                         colorscale='RdBu', opacity=0.8))  # Define los datos, la escala de colores y la opacidad

 #'Greys'
#'YlGnBu'
#'Greens'
#'YlOrRd'
#'Bluered'
#'RdBu'

# Configuración adicional del gráfico
fig.update_layout(scene=dict(xaxis_title='X',  # Actualiza el diseño del gráfico estableciendo los títulos de los ejes x, y y z
                             yaxis_title='Y',
                             zaxis_title='Z'),
                  title='3D Surface Plot - Synthetic Data')  # Establece el título del gráfico

# Mostrar el gráfico
fig.show()  # Muestra el gráfico interactivo en la pantalla
```

**Mayavi:**

Mayavi es una biblioteca específica para gráficos 3D. Es poderosa y permite crear visualizaciones 3D complejas. Se integra bien con NumPy y es ideal para aplicaciones científicas y de ingeniería. / se recomienda usarla en entornos de escritorio / En mi caso voy eligir nuestra amiga anaconda - **esta fórmula se utiliza para generar valores para una función sinusoidal tridimensional** / TAREA

```python
import numpy as np
from mayavi import mlab

# Crear datos tridimensionales con un rango más pequeño para estabilidad
x, y, z = np.mgrid[-3:3:50j, -3:3:50j, -3:3:50j]

# Asegurarse de que los valores no sean None ni NaN
values = np.sin(np.sqrt(x**2 + y**2 + z**2))
values = np.nan_to_num(values)

# Crear figura
mlab.figure(size=(800, 600), bgcolor=(1, 1, 1))

# Crear el volumen 3D
mlab.contour3d(x, y, z, values, contours=10, colormap='viridis')

# Ajustar vista
mlab.view(azimuth=45, elevation=60, distance=12)

# Mostrar
mlab.show()
```

##Calcular el volumen de un cilindro utilizando una integral simple. El volumen de un cilindro se puede calcular mediante la integral:

\[ V = \pi r^2 h \]

donde \( r \) es el radio de la base del cilindro y \( h \) es la altura.


```python
import numpy as np  # Importa la biblioteca NumPy y la renombra como 'np'
from scipy.integrate import quad  # Importa la función 'quad' del paquete scipy.integrate
from mayavi import mlab  # Importa la función 'mlab' del paquete Mayavi

# Parámetros del cilindro
radius = 2.0  # Define el radio del cilindro
height = 5.0  # Define la altura del cilindro

# Función que define el cilindro
def integrand(z):  # Define la función integrando para la integral
    return np.pi * radius**2  # Retorna el área de la sección transversal del cilindro (círculo) para cada valor de z

# Calcular la integral para obtener el volumen
result, error = quad(integrand, 0, height)  # Calcula la integral de la función integrando desde 0 hasta la altura para obtener el volumen del cilindro

# Imprimir el resultado del cálculo
print("Volumen calculado del cilindro:", result)  # Imprime el resultado del cálculo del volumen del cilindro

# Crear datos para la visualización
theta = np.linspace(0, 2 * np.pi, 100)  # Crea 100 puntos uniformemente espaciados en el rango de 0 a 2*pi para el ángulo theta
z = np.linspace(0, height, 100)  # Crea 100 puntos uniformemente espaciados en el rango de 0 a la altura para el eje z
theta, z = np.meshgrid(theta, z)  # Crea una cuadrícula 2D a partir de los puntos theta y z
x = radius * np.cos(theta)  # Calcula las coordenadas x de los puntos en la cuadrícula
y = radius * np.sin(theta)  # Calcula las coordenadas y de los puntos en la cuadrícula

# Crear la figura 3D
fig = mlab.figure(size=(800, 600), bgcolor=(1, 1, 1), fgcolor=(0, 0, 0))  # Crea una figura 3D con tamaño y colores definidos

# Visualizar el cilindro tridimensional
surface = mlab.mesh(x, y, z, colormap='viridis')  # Visualiza la superficie del cilindro en la figura 3D utilizando la cuadrícula de coordenadas

# Ajustar la vista
mlab.view(azimuth=45, elevation=60, distance=10)  # Ajusta la vista de la figura

# Agregar etiquetas y texto
mlab.xlabel('Eje X')  # Agrega una etiqueta al eje X
mlab.ylabel('Eje Y')  # Agrega una etiqueta al eje Y
mlab.zlabel('Eje Z')  # Agrega una etiqueta al eje Z

# Agregar el texto con el volumen calculado
mlab.text3d(radius, radius, height, f'Volumen: {result:.2f}', scale=(0.5, 0.5, 0.5), color=(0, 0, 0))  # Agrega un texto 3D al cilindro con el volumen calculado

# Mostrar la figura
mlab.show()  # Muestra la figura en pantalla
```

**Bokeh:**

Bokeh es una biblioteca que se centra en la creación de gráficos interactivos para la web. Puede manejar visualizaciones 3D a través de su extensión bokeh.plotting.

```python
from bokeh.plotting import figure, output_file, show, ColumnDataSource   # Importa funciones para crear figuras, mostrar gráficos y manejar datos en Bokeh
from bokeh.models import ColorBar, LinearColorMapper                     # Importa modelos para barra de color y mapeo de colores en Bokeh
from bokeh.palettes import Viridis256                                    # Importa la paleta de colores Viridis con 256 colores
import numpy as np                                                       # Importa NumPy para manejar arreglos numéricos

# Generamos datos aleatorios
x = np.random.rand(200) * 10  # Crea 200 valores aleatorios entre 0 y 1, luego los multiplica por 10 para tener un rango [0,10] en x
y = np.random.rand(200) * 10  # Crea 200 valores aleatorios entre 0 y 10 para y, igual que para x
z = np.sin(x) + np.cos(y)     # Calcula z como sin(x) + cos(y), generando un valor asociado a cada punto (x,y)

# Configuramos el mapeo de colores con base en z
mapper = LinearColorMapper(palette=Viridis256, low=z.min(), high=z.max())  # Crea un mapeador de color lineal que va desde el valor mínimo al máximo de z

# Creamos una fuente de datos para el gráfico
source = ColumnDataSource(data=dict(x=x, y=y, z=z))  # Crea un objeto ColumnDataSource con las columnas x, y, z

# Curareamos la fig
p = figure(title="Gráfico Interactivo con Bokeh",    # Establece el título del gráfico
           x_axis_label="Eje X",                     # Etiqueta del eje X
           y_axis_label="Eje Y",                     # Etiqueta del eje Y
           width=600, height=600)                    # Define el ancho y alto del gráfico en pixeles

# Dibujamos círculos en el gráfico, con color dependiente de z
p.scatter(x='x', y='y', source=source,               # Usa las columnas x e y del source para posicionar los círculos
         fill_color={'field': 'z', 'transform': mapper},  # El color de relleno se calcula según el valor z, mapeado por 'mapper'
         line_color='black', size=8, alpha=0.7)            # Contorno negro, tamaño de punto 8, transparencia 0.7

# Creamos una barra de color a la derecha que muestre la escala de z
color_bar = ColorBar(color_mapper=mapper,            # Barra de color usando el mismo mapper que los puntos
                     label_standoff=12,              # Distancia del texto de las etiquetas al color bar
                     width=8, location=(0,0))        # Ancho de la barra y ubicación
p.add_layout(color_bar, 'right')                     # Agregamos la barra de color al lado derecho del gráfico

# Definimos el archivo HTML de salida donde se guardará el gráfico
output_file("grafico_bokeh_sencillo.html")           # Crea o sobreescribe el archivo HTML con el gráfico

# Mostramos el gráfico en el navegador
show(p)  # Abre el gráfico en el navegador, interactivo (zoom, pan, etc.)
```

#crea un DataFrame con las características de las flores, asigna el nombre de las especies a cada muestra, y luego construye un gráfico interactivo de dispersión con Altair para visualizar la relación entre la longitud y el ancho del sépalo, coloreando los puntos según la especie y mostrando información relevante al pasar el cursor sobre ellos.

```python
import pandas as pd         # Importa la librería pandas para manejar dataframes
import altair as alt        # Importa Altair para crear visualizaciones interactivas
from sklearn.datasets import load_iris  # Importa la función para cargar el conjunto de datos Iris

# Cargar el conjunto de datos Iris
iris = load_iris()

# Crear un DataFrame con las características (columnas) del Iris
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Agregar una columna con los nombres de las especies, aprovechando el indexing vectorizado
df['species'] = iris.target_names[iris.target]

# Definir las características para los ejes del gráfico
x_feature = 'sepal length (cm)'  # Nombre de la característica para el eje X
y_feature = 'sepal width (cm)'   # Nombre de la característica para el eje Y

# Crear un gráfico de dispersión interactivo con Altair
chart = alt.Chart(df).mark_circle(size=60).encode(
    x=alt.X(x_feature, title='Longitud del Sépalo (cm)'),  # Define el eje X y su título
    y=alt.Y(y_feature, title='Ancho del Sépalo (cm)'),     # Define el eje Y y su título
    color=alt.Color('species', title='Especie'),           # Colorea los puntos según la especie
    tooltip=[x_feature, y_feature, 'species']              # Muestra info (x, y, especie) al pasar el cursor
).interactive(                                             # Permite zoom y desplazamiento
).properties(
    title='Gráfico Interactivo del Conjunto de Datos Iris', # Título del gráfico
    width=400,   # Ancho del gráfico
    height=300   # Alto del gráfico
)

# Guardar el gráfico como un archivo HTML interactivo
chart.save('grafico_interactivo_iris_altair.html')

# Mostrar el gráfico en el entorno interactivo (si corresponde)
chart
```

**Tenemos a nuestra amiga Plotly Express muy facil de usar y que muchos no conocen!!**

```python
import plotly.express as px
from sklearn.datasets import load_iris
import pandas as pd

# Cargar el conjunto de datos Iris
iris = load_iris()
df = px.data.iris()

# Agrupar por especie y calcular el promedio de 'sepal_length'
mean_df = df.groupby('species', as_index=False)['sepal_length'].mean()

# Crear el gráfico de barras con las medias
bar_chart = px.bar(mean_df, x='species', y='sepal_length', color='species',
                   title='Iris Dataset - Mean Sepal Length by Species',
                   labels={'sepal_length': 'Sepal Length', 'species': 'Species'})

bar_chart.show()
```

**La biblioteca Altair es una biblioteca de visualización declarativa en Python, diseñada para crear visualizaciones interactivas y atractivas de manera sencilla. Altair se basa en la gramática de visualización de Vega-Lite y proporciona una interfaz de alto nivel para crear gráficos sin tener que preocuparse demasiado por los detalles de bajo nivel.** /TAREA

```python
import altair as alt
from sklearn.datasets import load_iris
import pandas as pd

# Cargar el conjunto de datos Iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

# Crear un gráfico interactivo con Altair
scatter_plot = alt.Chart(df).mark_circle().encode(
    x='sepal length (cm)',
    y='sepal width (cm)',
    size='petal length (cm)',
    color='species',
    tooltip=['species', 'sepal length (cm)', 'sepal width (cm)', 'petal length (cm)']
).properties(
    title='Iris Dataset - Interactive Scatter Plot'
)

scatter_plot
```
