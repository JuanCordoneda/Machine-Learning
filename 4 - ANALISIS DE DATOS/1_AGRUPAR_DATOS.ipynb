Ordenar por columnas y limitar la cantidad de resultados
Otra forma de resolver el problema de encontrar el mayor y el menos es con el m茅todo sort_values. Este m茅todo puede recibir un valor 煤nico (nombre de columna) o una lista (con varias columnas) y un orden asc o desc. Por default el orden es asc.

Si combinamos el ordenamiento con el m茅todo head() para limitar la cantidad de resultados, podemos encontrar los N primeros.

Anexo: volviendo al tema de la vectorizaci贸n
驴Por qu茅 es tan importante trabajar con Pandas y no con funciones escritas por nosotros en Python nativo y que procesen los datos dentro de un for loop?

Por un lado est谩 la comodidad. Hay mucha funcionalidad que ya est谩 desarrollada en Pandas. Existen funciones que resuelven muchos de los problemas cl谩sicos de manipular datos: agrupar, sumarizar, sacar estad铆sticas, filtrar, etc. Pero adem谩s hay una raz贸n de performance.

Veamos una demostraci贸n de que vectorizar es m谩s eficiente. Vamos a crear dos listas de 1.000.000 de n煤meros aleatorios cada una y vamos a tratar de multiplicar elemento por elemento con pandas y sin pandas:

```python
lista1 = list(np.random.randint(1, 100, 1000000)) #lista de 1 mill贸n de n煤meros aleatorios entre 1 y 100
lista2 = list(np.random.randint(1, 100, 1000000))
```

```python
%%timeit #indica cuanto tarda en ejecutarse el c贸digo que sigue
for x,y in zip(lista1,lista2):
    x * y
```

```python
# Ahora probemos hacer lo mismo con dos series de Pandas

serie1 = pd.Series(lista1)
serie2 = pd.Series(lista2)
```

```python
#CON PANDAS ES MUCHO MS RPIDO
%%timeit
resultado = serie1 * serie2
```
